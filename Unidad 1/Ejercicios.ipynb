Ejercicio 2:
Realiza un programa que convierta un número de segundos introducidos por el usuario en horas, minutos y segundos

```python
a = int(input("Ingrese la cantidad de segundos: "))
horas = a // 3600
resto =  a % 3600
minutos = resto // 60
resto2 = resto % 60
segundos = resto2
print(f"La hora es {horas}H/{minutos}M/{segundos}S")
```

Ejercicio 3:
Crea un programa que calcule y muestre la raíz cuadrada de una serie de números introducidos por el usuario. El programa termina cuando el usuario introduce un número negativo

```python
import math
a = int(input("Ingrese un número: "))
while a >= 0:
    print(math.sqrt(a))
    a = int(input("Ingrese un número: "))
```

Ejercicio 4:
Crea un programa que solicite la cantidad de kilovatios hora (kWh) consumidos en un mes y calcule el costo total del servicio según la siguiente tabla: 

```python
a = int(input("Ingrese la cantidad de kWh: "))
if a <= 100:
    print(a * 200)
elif a <= 300:
    print(a * 300)
else:
    print(a * 400)
```
